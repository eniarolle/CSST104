# **1. Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

# **2. Preparing the Data**

```python
data = pd.read_csv('/content/Lesson_4_student_grades_dataset.csv')
data.head()
```

# **3. Visualizing  the Data**

1.  attendance - grades



```python

plt.scatter(data['Attendance_Percentage'], data['Grades'])
plt.xlabel('Attendance_Percentage')
plt.ylabel('Grades')
plt.title ('Effect of Attendance in Grades')
plt.show()
```

# **4. Preparing Data for Modeling**

```python
X = data[['Attendance_Percentage']]
Y = data[['Grades']]
```

# **5. Fitting the Linear Regression Model**

```python
model = LinearRegression()
model.fit(X,Y)
```

# **6. Getting the Model Coefficients**

```python
slope = model.coef_[0]

intercept = model.intercept_

print('Slope :' ,slope)
print('Intercept : ' ,intercept)
```

# **7. Visualizing the Linear Regression Line**

```python
plt.scatter(X,Y)
plt.plot (X,model.predict(X), color = 'red')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.title ('Linear Regression')
plt.show()
```

# **8. Making Prediction**

```python
new_X = [[75]]
predicted_Y = model.predict(new_X)
print(f'Predicted Value of: {predicted_Y}')
```
